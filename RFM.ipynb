```python
data = data[data["Price"] > 0]
data["date"]= pd.to_datetime(data["date"])
```

```python
data.info()
```

```python
from datetime import timedelta 
present  = data["date"].max() + timedelta(1)
```

```python
group = data.groupby("id_cliente")
```

```python
rfm = group.agg({"date": lambda x:(present-x.max()).days,
                "Price": lambda x:x.sum()})
```

```python
rfm
```

```python
s = data["id_cliente"].value_counts()
ss= s.sort_index()
rfm["frequency"] = ss
rfm
```

```python
rfm.columns = ['recency','monetary','frequency']
rfm
```

```python
rfm
```

```python
rfm['frequency'].value_counts()
```

```python
rfm["frequency_modified"] = rfm['frequency'].rank(method = 'first')
```

```python
rfm["frequency_modified"]
```

```python
rfm["frequency_modified"].sort_values()
```

```python
rfm['R_quartile'] = pd.qcut(rfm['recency'], 4,['1','2','3','4'])
rfm['M_quartile'] = pd.qcut(rfm['monetary'], 4,['4','3','2','1'])
rfm['f_quartile'] = pd.qcut(rfm['frequency_modified'], 4,['4','3','2','1'])
```

```python
rfm = rfm[['recency','monetary','frequency','frequency_modified','R_quartile','M_quartile','f_quartile']]
```
